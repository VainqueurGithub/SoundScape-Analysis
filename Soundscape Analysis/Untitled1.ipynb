```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import maad
from maad import sound, features, rois
from maad.util import (
    date_parser, plot_correlation_map,
    plot_features_map, plot_features, false_Color_Spectro
    )
```

```python
df = date_parser("Basamba/", dateformat='%Y%m%d_%H%M%S', verbose=True)
```

```python
df = df.reset_index()
df
```

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Hour'] = df['Date'].dt.hour
df['Minute'] = df['Date'].dt.minute
df['Second'] = df['Date'].dt.second
```

```python
df
```

```python
df_5 = df.loc[(df['Month']==6) & (df['Day'].isin([11]))]
df_5
```

```python
ACI_a = []
ADI_a = []
AEI_a = []
BI_a = []
NDSI_a = []
ratioBA_a = []
antroPh_a = []
bioPh_a = []
Ht_a = []
M_a = []
AR_a = []
ROItotal_a = []
ROIcover_a = []
```

```python
for i in df_5['file']:
    print(i)
```

```python
j=0
for i in df_5['file']:
    j+=1
    s, fs = sound.load(i) # Load the recording
    Sxx, tn, fn, ext = maad.sound.spectrogram (s, fs, mode='amplitude')  
    _, _ , ACI  = maad.features.acoustic_complexity_index(Sxx) # Computer the ACI
    ACI_a.append(round(ACI, 2))
      
    ADI  = maad.features.acoustic_diversity_index(Sxx,fn,fmax=10000, dB_threshold = -47) # Computer the ADI
    ADI_a.append(round(ADI, 2))

    AEI  = maad.features.acoustic_eveness_index(Sxx,fn,fmax=10000, dB_threshold = -47) # Computer the AEI
    AEI_a.append(round(AEI, 2))

    BI  = maad.features.bioacoustics_index(Sxx,fn,R_compatible=None) # Computer BI
    BI_a.append(round(BI, 2))
    
    Sxx_power, tn_p, fn_p, ext_p = maad.sound.spectrogram (s, fs)
    NDSI, ratioBA, antroPh, bioPh  = maad.features.soundscape_index(Sxx_power,fn_p,R_compatible=None)
    NDSI_a.append(round(NDSI, 2))
    ratioBA_a.append(round(ratioBA, 2))
    antroPh_a.append(round(antroPh, 2))
    bioPh_a.append(round(bioPh, 2))
    
    Ht = maad.features.temporal_entropy(s)
    M = maad.features.temporal_median(s)
    AR = maad.features.acoustic_richness_index([Ht],[M]) # Computer AR
    Ht_a.append(round(Ht, 2))
    M_a.append(round(M, 2))
    AR_a.append(AR)

    Sxx_noNoise= maad.sound.median_equalizer(Sxx_power) 
    Sxx_dB_noNoise = maad.util.power2dB(Sxx_noNoise)
    ROItotal, ROIcover = maad.features.region_of_interest_index(Sxx_dB_noNoise, tn_p, fn_p, display=False)
    ROItotal_a.append(round(ROItotal, 2))
    ROIcover_a.append(round(ROIcover, 2))
    print(j)
    
```

```python
df_5['ACI'] = ACI_a
df_5['ADI'] = ADI_a
df_5['AEI'] = AEI_a
df_5['BI'] = BI_a
df_5['NDSI'] = NDSI_a
df_5['ratioBA'] = ratioBA_a
df_5['antroPh'] = antroPh_a
df_5['bioPh'] = bioPh_a
df_5['Ht'] = Ht_a
df_5['M'] = M_a
df_5['AR'] = AR_a
df_5['ROItotal'] = ROItotal_a
df_5['ROIcover'] = ROIcover_a
```
