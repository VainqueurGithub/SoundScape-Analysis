```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df_5 = pd.read_csv('acoustic_indices_5_basamba.csv', sep=';')
df_6_1_5 = pd.read_csv('acoustic_indices_6_1_5_basambaoo.csv', sep=';')
#df_11 = pd.read_csv('avian_activity_11_basamba.csv', sep=';')
```

```python
#df_5['Date'] = pd.to_datetime(df_5['Date']).dt.date
df_5 = pd.concat([df_5, df_6_1_5], axis=0)
```

```python
df_5 = df_5.drop(columns=['file'])
df_5
```

```python
Day_group = []
for i in df_5['Hour']:
    if i ==0:
        Day_group.append('Night')
    if i ==1:
        Day_group.append('Night')
    if i ==2:
        Day_group.append('Night')
    if i ==3:
        Day_group.append('Night')
    if i ==4:
        Day_group.append('Night')
    if i ==5:
        Day_group.append('Dawn')
    if i ==6:
        Day_group.append('Dawn')
    if i ==7:
        Day_group.append('Day')
    if i ==8:
        Day_group.append('Day')
    if i ==9:
        Day_group.append('Day')
    if i ==10:
        Day_group.append('Day')
    if i ==11:
        Day_group.append('Day')
    if i ==12:
        Day_group.append('Day')
    if i ==13:
        Day_group.append('Day')
    if i ==14:
        Day_group.append('Day')
    if i ==15:
        Day_group.append('Day')
    if i ==16:
        Day_group.append('Day')
    if i ==17:
        Day_group.append('Dusk')
    if i ==18:
        Day_group.append('Dusk')
    if i ==19:
        Day_group.append('Night')
    if i ==20:
        Day_group.append('Night')
    if i ==21:
        Day_group.append('Night')
    if i ==22:
        Day_group.append('Night')
    if i ==23:
        Day_group.append('Night')
        
```

```python
df_5['Day_group'] = Day_group
df_5
```

```python
from statsmodels.graphics.tsaplots import plot_acf 
from statsmodels.graphics.tsaplots import plot_pacf
# Autocorrelation Plot 
plt.figure(figsize=(7,5)) 
plot_acf(df_5['ACI'], lags=50) 
plt.xlabel('Lags') 
plt.ylabel('Autocorrelation') 
plt.title('Autocorrelation Plot for ACI') 
plt.grid(True) 
plt.show()
```

```python
# PACF plot 
plt.figure(figsize=(7, 5)) 
plot_pacf(df_5['ACI'], lags=50) 
plt.xlabel('Lags') 
plt.ylabel('Partial Autocorrelation') 
plt.title('Partial Autocorrelation Function (PACF) Plot') 
plt.grid(True) 
plt.show() 
```

```python
# Extracting the month and year from the index of the above dataset "Monthly Sunspots" 
df_5['Date'] = df_5.index.hour 
  
# Grouping the data by month, calculating the average number of sunspots for each month 
hourly_average = df_5.groupby('Hour')['ACI'].mean() 
  
# Polar Plot theta (angle) and radii (length) settings 
theta = np.linspace(0, 2 * np.pi, len(hourly_average)) 
radii = hourly_average.values 
  
# Polar Plot 
plt.figure(figsize=(14, 5)) 
plt.polar(theta, radii) 
plt.title('Polar Plot of Monthly Average Sunspots') 
plt.xticks(range(1, 25), [ 
           '0h', '1h', '2h', '3h', '4h', '5h', '6h', '7h', '8h', '9h', '10h', '11h','12h', '13h', '14h', '15h', '16h', '17h', '18h', '19h', '20h', '21h', '22h', '23h']) 
  
# Set y-axis limit to accommodate the data 
plt.ylim(0, radii.max() + 10) 
plt.show() 
```

```python
# Extracting the month and year from the index of the above dataset "Monthly Sunspots" 
df_5['Date'] = df_5.index.month 
  
# Grouping the data by month, calculating the average number of sunspots for each month 
hourly_average = df_5.groupby('Month')['ACI'].mean() 
  
# Polar Plot theta (angle) and radii (length) settings 
theta = np.linspace(0, 2 * np.pi, len(hourly_average)) 
radii = hourly_average.values 
  
# Polar Plot 
plt.figure(figsize=(14, 5)) 
plt.polar(theta, radii) 
plt.title('Polar Plot of Monthly Average Sunspots') 
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 
                   'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']) 
  
# Set y-axis limit to accommodate the data 
plt.ylim(0, radii.max() + 10) 
plt.show() 
```

```python
# Moving Average Plot 
plt.figure(figsize=(15, 5)) 
values = df_5['ACI'] 
  
# 7-day moving average 
rolling_mean = values.rolling(window=7).mean() 
plt.plot(values, label='Original') 
plt.plot(rolling_mean, label='7-day Moving Average', color='red') 
plt.xlabel('Date') 
plt.ylabel('Value') 
plt.title('Moving Average Plot') 
plt.legend() 
plt.grid(True) 
plt.show() 
```
