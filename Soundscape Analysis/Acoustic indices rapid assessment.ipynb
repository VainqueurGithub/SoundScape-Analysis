CLUSTERING ANALYSIS 

```python
from sklearn.neighbors import NearestNeighbors
from sklearn import metrics
```

```python
nbrs = NearestNeighbors(n_neighbors = 4).fit(df_5[['ACI', 'ADI', 'AEI', 'BI', 'NDSI', 'Ht', 'M', 'ROIcover']])
```

```python
neigh_dist, neigh_ind = nbrs.kneighbors(df_5[['ACI', 'ADI', 'AEI', 'BI', 'NDSI', 'Ht', 'M', 'ROIcover']])
```

```python
sort_neigh_dist = np.sort(neigh_dist, axis = 0)
```

```python
k_dist = sort_neigh_dist[:, 3]
plt.plot(k_dist)
plt.ylabel("k-NN distance")
plt.xlabel("Sorted observations (3th NN)")
plt.show()
```

```python
from kneed import KneeLocator
kneedle = KneeLocator(x = range(1, len(neigh_dist)+1), y = k_dist, S = 1.0, 
                      curve = "concave", direction = "increasing", online=True)

# get the estimate of knee point
print(kneedle.knee_y)
```

```python
kneedle.plot_knee()
plt.show()
```
