```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import maad
from maad import sound, features, rois
from maad.util import (
    date_parser, plot_correlation_map,
    plot_features_map, plot_features, false_Color_Spectro
    )
```

```python
df = date_parser("Basamba/", dateformat='%Y%m%d_%H%M%S', verbose=True)
df = df.reset_index()
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Hour'] = df['Date'].dt.hour
df['Minute'] = df['Date'].dt.minute
df['Second'] = df['Date'].dt.second
df_5 = df.loc[(df['Month']==12) & (df['Day'].isin([23,24,25,26,27,28,29,30]))]
print(df_5.to_string())
```

```python
len(df_5)
```

```python
TFSD_a = []
j = 0
```

```python
for i in df_5['file']:
    j+=1
    s, fs = sound.load(i) # Load the recording
    Sxx_power,tn,fn,_ = maad.sound.spectrogram(s,fs)
    TFSD = maad.features.tfsd(Sxx_power,fn, tn, flim=(2000, 8000), log=False, mode='thirdOctave', display=False)
    TFSD_a.append(round(TFSD, 2))
    print(j)
```

```python
TFSD_b = []
k = 0
for i in df_5['file']:
    k+=1
    s, fs = sound.load(i) # Load the recording
    Sxx_power,tn,fn,_ = maad.sound.spectrogram(s,fs)
    TFSD = maad.features.tfsd(Sxx_power,fn, tn, flim=(0,1000), log=False, mode='thirdOctave', display=False)
    TFSD_b.append(round(TFSD, 2))
    print(k)
```

```python
df_5['TFSD_avian'] = TFSD_a
df_5['TFSD_human'] = TFSD_b
```

```python
df_5.to_csv('avian_activity_12_23_30_basamba.csv', sep=';', index=False)
```
